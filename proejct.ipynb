(i) 0 < g'(x) < 1

```python
def g1(x):
    f = e**(0.5*(x+2)) - 3
    return f
```

```python
x = linspace(-2.2,-1.9)
y =  e**(0.5*(x+2)) - 3
yy = x
plot(x,y,'-b', x, yy,'-r')
p0 = -2.2
for i in range(50):
    p1 = g1(p0)
    plot([p0,p0], [p0,p1],'g-')
    plot([p0,p1], [p1,p1],'-g')
    p0 = p1
print(p0)
```

```python
x = linspace(-2.2,-1.9)
y =  e**(0.5*(x+2)) - 3
yy = x
plot(x,y,'-b', x, yy,'-r')
h = 0.0001

g = g1

plot(x,y,'-b', x, yy,'-r')
x0 = -2.2
for i in range(50):
    gx0 = g(x0)
    gpx0 = (g(x0+h) - g(x0)) / h

    if (gpx0 > 0):
        h = -(abs(h))
    else:
        h = +(abs(h))

    if (gpx0 > 1) :
      if (gx0 > x0) :
        x1 = gx0 - 2 * abs(gx0 - x0)
      else :
        x1 = gx0 + 2 * abs(gx0 - x0)
    elif (1 > gpx0 and gpx0 > -1):
        x1 = gx0
    elif (-1 > gpx0) :
        raise NotImplemented

    plot([x0,x0], [x0,x1],'g-')
    plot([x0,x1], [x1,x1],'-g')
    x0 = x1
```

(iv) g'(x) < -1

```python
def g4(x):
    f = -e**(0.3 * (x+2)) - x 
    return f 
```

```python
x = linspace(-5,5)
y =  -e**(0.5 * (x+2)) - x
yy = x
plot(x,y,'-b', x, yy,'-r')
p0 = -0.5
for i in range(10):
    p1 = g4(p0)
    plot([p0,p0], [p0,p1],'g-')
    plot([p0,p1], [p1,p1],'-g')
    p0 = p1
print(p0)
```
